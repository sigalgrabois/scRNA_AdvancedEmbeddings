```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_variable_arrays_from_file(file_name):
    arrays = []
    with open(file_name, 'r') as file:
        for line in file:
            # Split the line into items, remove spaces, and convert to the appropriate type if needed
            array = line.strip().split(', ')
            # Here we assume that the contents are numeric and convert them to float.
            # If they are not numeric, remove the float conversion or adjust as necessary.
            arrays.append([float(item) for item in array])
    return arrays
```

```python
acc_data = read_variable_arrays_from_file('C:/Users/sigal/OneDrive - Bar Ilan University/master_degree/courses/SEMESTER_A/Unsupervised_Learning/NEW_PROJECT/acc.txt')
time_data = read_variable_arrays_from_file('C:/Users/sigal/OneDrive - Bar Ilan University/master_degree/courses/SEMESTER_A/Unsupervised_Learning/NEW_PROJECT/time.txt')
```

```python
acc_data_with_density = read_variable_arrays_from_file('C:/Users/sigal/OneDrive - Bar Ilan University/master_degree/courses/SEMESTER_A/Unsupervised_Learning/NEW_PROJECT/acc_with_density.txt')
time_data_with_density = read_variable_arrays_from_file('C:/Users/sigal/OneDrive - Bar Ilan University/master_degree/courses/SEMESTER_A/Unsupervised_Learning/NEW_PROJECT/time_with_density.txt')
```

```python
cluster_nums = range(1, 16) 
```

```python
plt.figure(figsize=(8, 6))
barWidth = 0.2
r1 = np.arange(len(cluster_nums))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

plt.bar(r1, acc_data[0], color='r', width=barWidth, edgecolor='grey', label='Original Data')
plt.bar(r2, acc_data[1], color='g', width=barWidth, edgecolor='grey', label='PCA')
plt.bar(r3, acc_data[2], color='y', width=barWidth, edgecolor='grey', label='Word2Vec')
plt.bar(r4, acc_data[3], color='b', width=barWidth, edgecolor='grey', label='UCE')

plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('k mean++ Accuracy of Different Methods for Different k Values')
plt.xticks([r + barWidth for r in range(len(cluster_nums))], cluster_nums)
# make the legend not on top of the bars
plt.legend(loc='upper left')
# add space at the top of the plot
plt.ylim(0, 1.1)

plt.show()
```
